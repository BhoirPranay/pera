int prime(int n)
{
  if(n<1){
    return 1;
  }
  else{
    for(int i=2;i<=n;i++)
      {
        if (n%i==0)
        {
          printf("it is not an prime number");
        }
        else{
          printf("it is  prime number");
        }
        return 0;
      }
  }
}

```python
// to find palidrome or not 
#include<stdio.h>
#include<string.h>
int main()
{ 
  int flag,len,i;
  char n[100];
  printf("enter the string :");
  scanf("%c",&n);
  len=strlen(n);
  for(i=0;i<len;i++)
    {
      if(n[i]!=n[len-i-1])
      {
        flag=1;
        break;
      }
      if(flag==1)
      {
        printf("it is palindrome");
      }
      else
      {
        printf("it is not palindrome");
      }
    }
  
  
}
```
